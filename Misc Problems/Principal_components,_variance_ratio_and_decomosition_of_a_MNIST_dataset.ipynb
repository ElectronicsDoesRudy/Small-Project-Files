```python
# Load Packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
np.set_printoptions(suppress=True)

# Import Data
df = pd.read_csv('https://confrecordings.ams3.digitaloceanspaces.com/mnist.csv')

# Prepare X and Y
Y = df.loc[:,"0"]
X = df.drop(['0'], axis=1)

print(df.shape)
print(df.head())
```

```python
#Import PCA Library
from sklearn.decomposition import PCA

#Load or Intialize PCA
pca = PCA(n_components=2)

#Fit the X data
principalComponents = pca.fit_transform(X)

#Make a dataframe and give columns to the name
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])

#Make a final dataframe with concatenation of Y column
finalDf = pd.concat([principalDf,Y], axis = 1)

#print the top 5 rows
print(finalDf.head())

#print the variance ratio
print(pca.explained_variance_ratio_)
```

```python
#Plot PCA1 and PCA2

#import library
import seaborn as sns

#ploy scatterplot
sns.scatterplot(data = finalDf, x = 'principal component 1', y = 'principal component 2', hue = '0')

#Show the plot
plt.show()
```
