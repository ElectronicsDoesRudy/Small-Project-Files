```python
#import library
import pandas as pd

#read dataset
data=pd.read_csv('https://confrecordings.ams3.digitaloceanspaces.com/data_1612.csv')

#inspect the data
print(data.head())
print(data.shape)
print(data.columns)
print(data.dtypes)
print(data.info())

from sklearn import preproccessing
label_encoder = preprocessing.LabelEncoder()

data['crop']= label_encoder.fit_transform(data['crop'])
data['moisture']= label_encoder.fit_transform(data['moisture'])
data['temp']= label_encoder.fit_transform(data['temp'])
data['pump']= label_encoder.fit_transform(data['pump'])
print(data.dtypes())

X=data[['crop','moisture','temp']]
y=data['pump']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=1)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```
