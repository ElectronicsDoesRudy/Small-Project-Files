```python
#import the libraries

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np

#storing the dataset in the variable data

housing=pd.read_csv('https://confrecordings.ams3.digitaloceanspaces.com/house.csv')

#Splitting the dataset into two: train and test x_train and x_test are the independent variables of the train and test dataset y_train and y_test are the dependent variables of the train and test dataset

x_train,x_test,y_train,y_test = train_test_split(housing["area"],housing['price'], test_size = 0.20)

#Calling the LinearRegression class and storing it in the variable reg

reg = LinearRegression()

#Using the fit function to train the model reg

reg.fit(np.array(x_train).reshape(-1,1),y_train)

#Predicting a value using the reg model by predict function

pred = reg.predict([[69]])
print(pred)

#Predict for all x_test value

print(reg.predict(np.array(x_test).reshape(-1,1)))

#Import all libraries
import seaborn as sns
import matplotlib.pyplot as plt

#plot the graph
sns.Implot(x="area",y="price",data=housing)
plt.show()
```
