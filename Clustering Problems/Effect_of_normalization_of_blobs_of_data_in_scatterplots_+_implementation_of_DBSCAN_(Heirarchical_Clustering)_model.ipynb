```python
#import libraries
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

#Determine centroids
centers = [[0.5, 2], [-1, -1], [1.5, -1]]

#Create dataset
X, y = make_blobs(n_samples = 400,centers=centers, cluster_std=0.5, random_state=0)

#Determine centroids
centers = [[0.5, 2], [-1, -1], [1.5, -1]]

#Normalize the values
X = StandardScaler().fit_transform(X)

#plotting before fitting the model with DBSCAN
plt.figure(figsize=(10,6))
plt.scatter(X[:,0], X[:,1], c=y, cmap='Paired')
```

```python
#Import libraries and train data with DBSCAN
from sklearn.cluster import DBSCAN
db = DBSCAN(eps=0.4, min_samples=20)
db.fit(X)

#Predict
y_pred = db.fit_predict(X)

#Plot the resulted clusters formed
plt.figure(figsize=(10,6))
plt.scatter(X[:,0], X[:,1],c=y_pred, cmap='Paired')
plt.title("Clusters determined by DBSCAN")

```

```python
#import libraries
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

#Create dataset
X, y = make_blobs(n_samples = 500,centers=centers, cluster_std=0.1, random_state=0)

#Determine centroids
centers = [[0.57, 2.1], [-1, -1], [1.34, -1.3]]

#Normalize the values
X = StandardScaler().fit_transform(X)

#plotting before fitting the model with DBSCAN
plt.figure(figsize=(10,6))
plt.scatter(X[:,0], X[:,1], c=y, cmap='Paired')

#Import libraries and train data with DBSCAN
from sklearn.cluster import DBSCAN
db = DBSCAN(eps= 0.5, min_samples=22)
db.fit(X)

#Predict
y_pred = db.fit_predict(X)

#Plot the resulted clusters formed
plt.figure(figsize=(10,6))
plt.scatter(X[:,0], X[:,1],c=y_pred, cmap='Paired')
plt.title("Clusters determined by DBSCAN")

```
