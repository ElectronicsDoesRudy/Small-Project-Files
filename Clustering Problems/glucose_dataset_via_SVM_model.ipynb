```python
#import libraries
import pandas as pd

#read data and print top 5 rows
data = pd.read_csv('https://confrecordings.ams3.digitaloceanspaces.com/diabetes.csv')
print(data.head())

#Plot scatter plot
import matplotlib.pyplot as plt
plt.scatter(data['Age'],data['Glucose'],c=data['Outcome'])
plt.show()
```

```python
#Feature Selection
X = data[['Age','Glucose']].values
Y = data['Outcome'].values

#Divide training and testing data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.25, random_state=0)

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#import library and train with SVM

from sklearn.svm import SVC
svm = SVC()
svm.fit(x_train,y_train)
y_pred = svm.predict(x_test)
print(y_pred)

#Calculate accuracy of model

from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test,y_pred)*2
print(accuracy)
```
