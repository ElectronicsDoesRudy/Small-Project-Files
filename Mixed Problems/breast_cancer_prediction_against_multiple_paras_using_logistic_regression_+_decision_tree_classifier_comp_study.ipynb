```python
#import libraries
import pandas as pd

#Read csv file
df = pd.read_csv('https://confrecordings.ams3.digitaloceanspaces.com/breast-cancer-wisconsin.data',header=None)
df1 = pd.read_csv("https://confrecordings.ams3.digitaloceanspaces.com/breast-cancer-wisconsin.data")

print(df1.head)
print(df1.shape)
print(df1.dtypes)
#Add column names
df.columns=['a','b','c','d','e','f','g','h','i','j','class']

#Perform EDA
print(df.head(5))
print(df.shape)
print(df.isnull().sum())
print(df.dtypes)

#Convert 2 into 0 and 4 into 1 present in last column
df.iloc[:,10].replace(2, 0,inplace=True)
df.iloc[:,10].replace(4, 1,inplace=True)

#Eliminates those rows that contains ?
df.drop("g", inplace = True, axis = 1)

#Convert whole dataframe type into
df = df.astype(float)

#Choose your independent and dependent column and store in x and y variable
y = df['class']
X = df.drop("class", axis = 1)

#You can check value in x and y by printing
print(X)
print(y)

#Divide data into 80% training and 20% testing data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=0)
#Check shape of all 4 train and test variables
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)

#Train using decision tree model and calculate its  score/accuracy
from sklearn.metrics import roc_auc_score
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(criterion='gini', min_samples_split=200,min_samples_leaf=100, max_depth=3)
dt.fit(X_train, y_train)
y_pred = dt.predict(X_test)
print(roc_auc_score(y_test, y_pred))

#Train using logistic regression and calculate its score/accuracy
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(C=0.001, random_state=1)
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
print(roc_auc_score(y_test, y_pred))
```
