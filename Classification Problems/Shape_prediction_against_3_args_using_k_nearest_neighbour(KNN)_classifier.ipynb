```python
#importing the library

import pandas as pd
dataset = pd.read_csv('https://confrecordings.ams3.digitaloceanspaces.com/DL_8_Shape%20prediction.csv')


#dataset inspection

print(dataset.columns)
print(dataset.isnull().sum())
print(dataset.shape)
print(dataset.head())
print(dataset['Shape'].value_counts())

```

```python
#import library
from sklearn import preprocessing

#creating labelEncoder
le = preprocessing.LabelEncoder()

# Converting string labels into numbers.
colour_encoded=le.fit_transform(dataset['Colour'])
print(colour_encoded)
outline_encoded=le.fit_transform(dataset['Outline'])
print(outline_encoded)
circle_encoded=le.fit_transform(dataset['Circle inside Shape'])
print(circle_encoded)
label=le.fit_transform(dataset['Shape'])
print(label)

#combining
features=list(zip(circle_encoded,outline_encoded,colour_encoded))

```

```python
#import library
from sklearn.neighbors import KNeighborsClassifier

#load K-NN
model = KNeighborsClassifier(n_neighbors=3)

# Train the model using the training sets
model.fit(features,label)

#Predict Output
predicted= model.predict([[0,2,0]])
print(predicted)
```
