```python
#import the libraries

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import pandas as pd
#storing the dataset in the variable data

cp=pd.read_csv('https://confrecordings.ams3.digitaloceanspaces.com/CarPrice_9654.csv')

#print top 5 rows of data
print(cp.head())

#Splitting the dataset into two: train and test x_train and x_test are the independent variables of the train and test dataset y_train and y_test are the dependent variables of the train and test dataset

x_train,y_train,x_test,y_test=train_test_split(cp[["enginesize",'horsepower']],cp["price"],test_size=0.25)

#Calling the LinearRegression class and storing it in the variable reg

reg = LinearRegression()

#Using the fit function to train the model reg

reg.fit(x_train,y_train)

#Predicting a value using the reg model by predict function

pred=reg.predict([[130,111]])
print(pred)

#Predict for all x_test value

print(reg.predict(x_test))
```

```python
#import library
import matplotlib.pyplot as plt
plt.figure(1,figsize=(15,6))
fig,ax = plt.subplots()

#plot enginesize vs price
ax.scatter(cp.enginesize ,cp.price, color="red", marker="o")

# set x-axis label
ax.set_xlabel("enginesize",fontsize=18)

# set y-axis label
ax.set_ylabel("price",color="red",fontsize=18)

#Make a twin y axis
ax2=ax.twiny()

# plot horsepower vs price
ax2.scatter(cp.horsepower,cp.price,color="blue",marker="o")

#set twi x-axis label
ax2.set_xlabel("horsepower",color="green",fontsize=18)

#show the plot
plt.show()
```
