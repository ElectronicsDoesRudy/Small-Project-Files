```python
# Importing the pandas library as pd
import pandas as pd

# Importing the dataset
data= pd.read_csv('https://confrecordings.ams3.digitaloceanspaces.com/Position_Salaries.csv')

#Printing top 5 rows of the dataset
print(data.head(5))

#Storing the independent column in X and the dependent column in y variable respectively
X = data.iloc[:, 1:2].values
Y = data.iloc[:, 2].values
X = X.reshape(-1,1)

#Printing X and y variables
print(X)
print(Y)

# Training the Random Forest Regression model on the whole dataset
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 100)
regressor.fit(X,Y)

# Predicting a new result
print(regressor.predict([[6.5]]))

#import library
import matplotlib.pyplot as plt

#Make line plot
plt.plot(X,Y, color = 'red')

#Make scatter plot
plt.scatter(3.5, regressor.predict([[3.5]]), color = 'blue')

#Add title , x-axis label & y-axis label
plt.title('Truth or Bluff (Random Forest Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()


```
