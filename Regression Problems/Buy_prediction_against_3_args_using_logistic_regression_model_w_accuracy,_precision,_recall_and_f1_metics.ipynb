```python
#write your code here

#import library
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score

y_pred = [0, 2, 2, 3,4,5,2,3,6,7,4,4,5,2,1,5]
y_true = [0, 2, 2, 3,4,5,2,2,5,6,4,4,5,4,1,5]

print('Accuracy:',round(accuracy_score(y_true, y_pred),2))

print('Precision:',round(precision_score(y_true, y_pred, average= 'weighted'),2))

print('Recall',round(recall_score(y_true, y_pred, average= 'weighted'),2))

print('F1 measure',round(f1_score(y_true, y_pred, average= 'weighted'),2))

```

```python
#write your code here

import pandas as pd
from sklearn.model_selection import train_test_split

data = pd.read_excel("https://confrecordings.ams3.digitaloceanspaces.com/Purchase_5239.xlsx")

daykey = {"Weekday":0,"Weekend":1,"Holiday":2}

data['Day']=data['Day'].map(daykey)

daykey= {'No':0, 'Yes':1}

data['Discount']=data['Discount'].map(daykey)

data['Free Delivery']=data['Free Delivery'].map(daykey)

data['Purchase']=data['Purchase'].map(daykey)

print(data.head())

X = data[["Day","Discount","Free Delivery"]]
y = data["Purchase"]

X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 0.2)

from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()

lr.fit(X_train,y_train)

y_pred = lr.predict(X_test)

print(y_pred)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score

print('Accuracy:',round(accuracy_score(y_test, y_pred),2))

print('Precision:',round(precision_score(y_test, y_pred, average= 'weighted'),2))

print('Recall',round(recall_score(y_test, y_pred, average= 'weighted'),2))

print('F1 measure',round(f1_score(y_test, y_pred, average= 'weighted'),2))

```
