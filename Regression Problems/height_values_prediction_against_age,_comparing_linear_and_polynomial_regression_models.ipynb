```python
#Importing the libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Importing the dataset
dataset = pd.read_csv('https://confrecordings.ams3.digitaloceanspaces.com/HeightVsWeight_3155.csv')

#Inspect the data
print(dataset.head())
print(dataset.columns)
print(dataset.dtypes)
print(dataset.shape)

#Making the first column as the independent variable and storing it in X
X = dataset.iloc[:,0].values

#Making the second column as the dependent variable and storing it in y
Y = dataset.iloc[:,1].values

#Converting X into a 2D array
X = X.reshape(-1,1)

# Training the Linear Regression model on the whole dataset
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,Y)

#Predicting a value using the trained model
prediction = lin_reg.predict([[56]])
print(prediction)

from sklearn.linear_model import LinearRegression

# Training the Polynomial Regression model on the whole dataset
from sklearn.preprocessing import PolynomialFeatures
features = PolynomialFeatures(degree = 4)
X_poly = features.fit_transform(X)
poly_reg = LinearRegression()
poly_reg.fit(X_poly,Y)


#Predicting a value using the trained model
prediction = poly_reg.predict(features.fit_transform([[56]]))
print(prediction)

# Visualising the Linear Regression results
plt.scatter(X,Y, color = 'red')
plt.plot(X, lin_reg.predict(X), color = 'blue')
plt.title("age versus hight graph")
plt.xlabel('Age')
plt.ylabel('Height')
plt.show()

# Visualising the Polynomial Regression results
plt.scatter(X, Y, color = 'red')
plt.plot(X, poly_reg.predict(features.fit_transform(X)), color = 'blue')
plt.title("poly nomia regression")
plt.xlabel('Age')
plt.ylabel("Height")
plt.show()



```
