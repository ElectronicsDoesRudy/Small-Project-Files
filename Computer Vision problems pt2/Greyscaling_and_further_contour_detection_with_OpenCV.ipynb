```python
import cv2
im=cv2.imread('Donut_Shapes.png')
cv2_imshow(im)
imgray=cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)
cv2_imshow(imgray)
mean=imgray.mean()
ret, thresh =cv2.threshold(imgray ,mean , 255 , 0)
cv2_imshow(thresh)
contours, hierarchy = cv2.findContours(thresh , cv2.RETR_EXTERNAL , cv2.CHAIN_APPROX_NONE)
cv2.drawContours(im,contours,contourIdx=-1,color=(255,0,0), thickness=2,lineType=cv2.LINE_AA)
cv2_imshow(im)
contours, hierarchy = cv2.findContours(thresh , cv2.RETR_TREE , cv2.CHAIN_APPROX_SIMPLE)
cv2.drawContours(im,contours,contourIdx=-1,color=(0,255,0), thickness=2,lineType=cv2.LINE_AA)
cv2_imshow(im)



```

```python
import cv2
img=cv2.imread('Shapes1.png')
cv2_imshow(img)
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
mean=gray.mean()
ret, thresh =cv2.threshold(gray ,mean , 255 , 0)
cv2_imshow(thresh)
contours, hierarchy = cv2.findContours(thresh , cv2.RETR_EXTERNAL , cv2.CHAIN_APPROX_NONE)
cv2.drawContours(img,contours,contourIdx=-1,color=(255,0,0), thickness=2,lineType=cv2.LINE_AA)
cv2_imshow(img)
```

```python
i=0
for contour in contours:
    if i == 0:
       i = 1
       continue
 # cv2.approxPloyDP() function to approximate the shape
    approx = cv2.approxPolyDP(contour, 0.01 * cv2.arcLength(contour, True), True)
                  # using drawContours() function
    cv2.drawContours(img, [contour], 0, (0, 0, 255), 5)
                      # finding center point of shape
    M = cv2.moments(contour)
    if M['m00'] != 0.0:
       x = int(M['m10']/M['m00'])
       y = int(M['m01']/M['m00'])
                                    # drawing a circle of radius 5 at center of each shape like we have done for shape who have three edges similarily we need to check for other shapes like quadrilateral,pentagon , circle etc.
    if len(approx) == 3:
       img = cv2.circle(img, (x,y) ,5, (255,255,255) ,-1)
       shape = "Triangle"
    elif len(approx) == 4:
         shape = "Quadrilateral"
    elif len(approx) > 4:
         shape = "Circle"
    else:
         shape = "Unknown"
                                                      # Display the final image with detected shapes and their centers
    cv2_imshow(img)




```

```python
from google.colab.patches import cv2_imshow
import cv2
def detect_contours(case):
    im = cv2.imread('ABC_Image.jpg')
    imgray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
    _, thresh = cv2.threshold(imgray, 127, 255, cv2.THRESH_BINARY)
    if case == 1:
       contours, hierarchy = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    elif case == 2:
       contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    elif case == 3:
       contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    for cnt in contours:
       x, y, w, h = cv2.boundingRect(cnt)
       cv2.rectangle(im, (x, y), (x+w, y+h), (0, 255, 0), 3)
       cv2.imshow("Detected Image", im)



detect_contours(1)
detect_contours(2)
detect_contours(3)
```
